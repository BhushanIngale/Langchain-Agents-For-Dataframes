# 1. Libraries & Set-up

```python
# Libraries:

import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np

import os
os.system("pip install opencv-python")
import cv2
```

```python
# Initialising Pre-trained model:

model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

# 2. Preprocess & Load Images

```python
# This is to Mount Google Drive since our Content & Style Image is in the Drive

#import os
#from google.colab import drive
#drive.mount('/content/drive')
```

```python
# Changing the path where Image is stored:

#os.chdir('/content/drive/My Drive/Colab Notebooks/Neural-Style-Transfer')
```

```python
# Creating a function to load and process the images:

def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
# Assigning images to variables with load_image function:

content_image = load_image('gateway.jpg')
style_image = load_image('starrynight.jfif')
```

# 3. Plotting Input Images

```python
# Shape of the image tensor:

content_image.shape
```

```python
# Plotting the style image:

plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
# Plotting the content image:

plt.imshow(np.squeeze(content_image))
plt.show()
```

# 4. Stylize Content Image

```python
# Passing the content image to our pre-trained model:

stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
# Plotting the new Stylised Image:

plt.imshow(np.squeeze(stylized_image))
plt.show()
```

```python
# Saving the image in local directory:

cv2.imwrite('generated_img.jpg', cv2.cvtColor(np.squeeze(stylized_image)*255, cv2.COLOR_BGR2RGB))
```
